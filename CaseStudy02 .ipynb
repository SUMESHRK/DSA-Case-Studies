# Find out how many males and females participated in the test.


```python
import pandas as pd
data_sdt=pd.read_csv('studentsPerformance.csv')
data_sdt['gender'].value_counts()
```

# What do you think about the students' parental level of education?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize =(15,5))
sns.countplot(x = data_sdt['parental level of education'], palette = 'RdBu')
plt.title('Comparison of Parental Education', fontweight = 30, fontsize = 20)
plt.yticks(fontname="Arial", fontsize=14)
plt.xticks(fontname="Arial", fontsize=14)
plt.xlabel("Degree",fontname="Arial", fontsize=16)
plt.ylabel("Count",fontname="Arial", fontsize=16)
plt.show()
```

# Average scores for math, reading and writing based on Gender

```python
gndr_math = data_sdt.groupby('gender')['math score'].mean()
#gndr_math
gndr_rs = data_sdt.groupby('gender')['reading score'].mean()
gndr_ws = data_sdt.groupby('gender')['writing score'].mean()

dtmath = pd.DataFrame({'Gender':['Female', 'Male'],
        'Score':[gndr_math[0], gndr_math[1]]})
#dtmath
dtrs = pd.DataFrame({'Gender':['Female', 'Male'],
        'Score':[gndr_rs[0], gndr_rs[1]]})

dtws = pd.DataFrame({'Gender':['Female', 'Male'],
        'Score':[gndr_ws[0], gndr_ws[1]]})

plt.figure(figsize=(20,10),facecolor='w')
 
plt.subplot(1,3,1) 
sns.barplot(data=dtmath, x="Gender", y="Score")
plt.title("Maths Score Female v/s Male",size=16)
plt.yticks(fontname="Arial", fontsize=14)
plt.xticks(fontname="Arial", fontsize=14)
plt.xlabel("Gender",fontname="Arial", fontsize=16)
plt.ylabel("Score of Student",fontname="Arial", fontsize=16)

plt.subplot(1,3,2) 
sns.barplot(data=dtrs, x="Gender", y="Score")
plt.title("Reading Score Female v/s Male",size=16)
plt.yticks(fontname="Arial", fontsize=14)
plt.xticks(fontname="Arial", fontsize=14)
plt.xlabel("Gender",fontname="Arial", fontsize=16)
plt.ylabel("Score of Student",fontname="Arial", fontsize=16)

plt.subplot(1,3,3) 
sns.barplot(data=dtws, x="Gender", y="Score")
plt.title("Writing Score Female v/s Male",size=16)
plt.yticks(fontname="Arial", fontsize=14)
plt.xticks(fontname="Arial", fontsize=14)
plt.xlabel("Gender",fontname="Arial", fontsize=16)
plt.ylabel("Score of Student",fontname="Arial", fontsize=16)
```
