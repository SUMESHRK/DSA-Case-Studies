# Check whether all the zones generate the same amount of sales

```python
import scipy.stats as stats
fval,pval = stats.f_oneway(sldata['Zone - A'],sldata['Zone - B'],sldata['Zone - C'], sldata['Zone - D'])
#print(fvalue,pvalue)

alpha = 0.05  # set the significance level
if pval > alpha:
    print("Accept the null hypothesis")
    print("Same amount of sale was generated in all zones.")
elif pval < alpha:
    print("Reject the null hypothesis")
    print("Sales amount was different in atleast one zone.")
```
